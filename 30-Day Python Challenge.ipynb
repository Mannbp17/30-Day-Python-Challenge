```python
# DAY 19

import json
import re
from collections import Counter


# Exercise 1
# Count number of lines and words in a file

def count_lines_words(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        num_lines = len(lines)
        num_words = sum(len(line.split()) for line in lines)
    return num_lines, num_words


print(count_lines_words('./data/obama_speech.txt'))
print(count_lines_words('./data/michelle_obama_speech.txt'))
print(count_lines_words('./data/donald_speech.txt'))
print(count_lines_words('./data/melina_trump_speech.txt'))


# Exercise 2
# Most spoken languages

def most_spoken_languages(filename, top):
    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

    languages = []

    for country in data:
        languages.extend(country['languages'])

    count = Counter(languages)
    return count.most_common(top)


print(most_spoken_languages('./data/countries_data.json', 10))
print(most_spoken_languages('./data/countries_data.json', 3))


# Exercise 3
# Most populated countries

def most_populated_countries(filename, top):
    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

    countries = sorted(data, key=lambda x: x['population'], reverse=True)

    result = []
    for country in countries[:top]:
        result.append({
            'country': country['name'],
            'population': country['population']
        })

    return result


print(most_populated_countries('./data/countries_data.json', 10))
print(most_populated_countries('./data/countries_data.json', 3))


# Exercise 4
# Extract email addresses

def extract_emails(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()

    emails = re.findall(r'\S+@\S+', text)
    return emails


print(extract_emails('./data/email_exchange_big.txt'))


# Exercise 5
# Find most common words in a text or file

def find_most_common_words(source, top):
    if source.endswith('.txt'):
        with open(source, 'r', encoding='utf-8') as f:
            text = f.read()
    else:
        text = source

    words = re.findall(r'\b\w+\b', text.lower())
    count = Counter(words)
    return count.most_common(top)


print(find_most_common_words('sample.txt', 10))
print(find_most_common_words('sample.txt', 5))


# Exercise 6
# Most frequent words in speeches

print(find_most_common_words('./data/obama_speech.txt', 10))
print(find_most_common_words('./data/michelle_obama_speech.txt', 10))
print(find_most_common_words('./data/donald_speech.txt', 10))
print(find_most_common_words('./data/melina_trump_speech.txt', 10))


# Exercise 7
# Clean text

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text


# Remove stop words

def remove_stop_words(text, stop_words):
    words = text.split()
    filtered = [word for word in words if word not in stop_words]
    return filtered


# Check similarity between two texts

def check_text_similarity(file1, file2, stop_words_file):

    with open(file1, 'r', encoding='utf-8') as f:
        text1 = f.read()

    with open(file2, 'r', encoding='utf-8') as f:
        text2 = f.read()

    with open(stop_words_file, 'r', encoding='utf-8') as f:
        stop_words = f.read().split()

    text1 = clean_text(text1)
    text2 = clean_text(text2)

    words1 = set(remove_stop_words(text1, stop_words))
    words2 = set(remove_stop_words(text2, stop_words))

    similarity = len(words1.intersection(words2)) / len(words1.union(words2))

    return similarity


print(check_text_similarity(
    './data/michelle_obama_speech.txt',
    './data/melina_trump_speech.txt',
    './data/stop_words.txt'
))


# Exercise 8
# 10 most repeated words in Romeo and Juliet

print(find_most_common_words('./data/romeo_and_juliet.txt', 10))


# Exercise 9
# Hacker news analysis

def hacker_news_analysis(filename):

    python_count = 0
    javascript_count = 0
    java_count = 0

    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            if re.search(r'\b[Pp]ython\b', line):
                python_count += 1

            if re.search(r'JavaScript|javascript|Javascript', line):
                javascript_count += 1

            if re.search(r'\bJava\b', line) and not re.search(r'JavaScript', line):
                java_count += 1

    return python_count, javascript_count, java_count


print(hacker_news_analysis('./data/hacker_news.csv'))
```

```python
# DAY 21

# Exercise Level 1

class Statistics:

    def __init__(self, data):
        self.data = data

    def count(self):
        return len(self.data)

    def sum(self):
        return sum(self.data)

    def min(self):
        return min(self.data)

    def max(self):
        return max(self.data)

    def range(self):
        return self.max() - self.min()

    def mean(self):
        return round(self.sum() / self.count())

    def median(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        mid = n // 2

        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        else:
            return sorted_data[mid]

    def mode(self):
        freq = {}
        for num in self.data:
            freq[num] = freq.get(num, 0) + 1

        mode_value = max(freq, key=freq.get)
        return {'mode': mode_value, 'count': freq[mode_value]}

    def var(self):
        mean = self.mean()
        variance = sum((x - mean) ** 2 for x in self.data) / len(self.data)
        return round(variance, 1)

    def std(self):
        return round(self.var() ** 0.5, 1)

    def freq_dist(self):
        freq = {}
        for num in self.data:
            freq[num] = freq.get(num, 0) + 1

        total = len(self.data)
        distribution = []

        for value, count in freq.items():
            percent = (count / total) * 100
            distribution.append((round(percent, 1), value))

        distribution.sort(reverse=True)
        return distribution

    def describe(self):
        print("Count:", self.count())
        print("Sum:", self.sum())
        print("Min:", self.min())
        print("Max:", self.max())
        print("Range:", self.range())
        print("Mean:", self.mean())
        print("Median:", self.median())
        print("Mode:", (self.mode()['mode'], self.mode()['count']))
        print("Variance:", self.var())
        print("Standard Deviation:", self.std())
        print("Frequency Distribution:", self.freq_dist())


ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]

data = Statistics(ages)

print('Count:', data.count())
print('Sum:', data.sum())
print('Min:', data.min())
print('Max:', data.max())
print('Range:', data.range())
print('Mean:', data.mean())
print('Median:', data.median())
print('Mode:', data.mode())
print('Standard Deviation:', data.std())
print('Variance:', data.var())
print('Frequency Distribution:', data.freq_dist())

print()
data.describe()


# Exercise Level 2

class PersonAccount:

    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, description, amount):
        self.incomes[description] = amount

    def add_expense(self, description, amount):
        self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        print("Account Holder:", self.firstname, self.lastname)
        print("Incomes:", self.incomes)
        print("Expenses:", self.expenses)
        print("Total Income:", self.total_income())
        print("Total Expense:", self.total_expense())
        print("Balance:", self.account_balance())


person = PersonAccount("John", "Doe")

person.add_income("Salary", 5000)
person.add_income("Freelance", 1200)

person.add_expense("Rent", 1500)
person.add_expense("Food", 400)

person.account_info()
```

```python
# DAY 25 
import pandas as pd

# 1. Read the CSV file
df = pd.read_csv('data/hacker_news.csv')

# 2. Get first five rows
print(df.head())

# 3. Get last five rows
print(df.tail())

# 4. Get the 'title' column as a Series
titles = df['title']
print(titles)

# 5. Count number of rows and columns
print('Shape:', df.shape)

# 6. Filter titles containing 'Python'
python_titles = df[df['title'].str.contains('Python', case=False, na=False)]
print('Python titles:')
print(python_titles)

# 7. Filter titles containing 'JavaScript'
js_titles = df[df['title'].str.contains('JavaScript', case=False, na=False)]
print('JavaScript titles:')
print(js_titles)

# 8. Explore the data
print(df.info())      # info about columns and data types
print(df.describe())  # basic statistics for numeric columns
print(df.isnull().sum()) # check missing values
```

```python
# DAY 26
# app.py
from flask import Flask, render_template, request, redirect, url_for
import os
from collections import Counter

app = Flask(__name__)
app.config['SEND_FILE_MAX_AGE_DEFAULT'] = 0  # stop caching static files

# Home route
@app.route('/')
def home():
    techs = ['HTML', 'CSS', 'Flask', 'Python']
    name = '30 Days Of Python Programming'
    return render_template('home.html', techs=techs, name=name, title='Home')

# About route
@app.route('/about')
def about():
    name = '30 Days Of Python Programming'
    return render_template('about.html', name=name, title='About Us')

# Post route (text analyzer)
@app.route('/post', methods=['GET','POST'])
def post():
    name = 'Text Analyzer'
    if request.method == 'GET':
        return render_template('post.html', name=name, title=name)
    if request.method == 'POST':
        content = request.form['content']
        # Analyze text
        words = content.split()
        num_words = len(words)
        num_chars = len(content)
        freq_words = dict(Counter(words).most_common(5))  # top 5 words
        return render_template('result.html', content=content,
                               num_words=num_words,
                               num_chars=num_chars,
                               freq_words=freq_words,
                               title='Result')

# Result route
@app.route('/result')
def result():
    return render_template('result.html')

if __name__ == '__main__':
    port = int(os.environ.get("PORT", 5000))
    app.run(debug=True, host='0.0.0.0', port=port)
```
